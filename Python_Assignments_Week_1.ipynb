```python
"""
Diamond Pattern:
1. Write a program to print a diamond shape using nested loops

    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *

"""

N = 5
for i in range(N):
  for j in range(N-i-1):
    print(' ', end='')
  for k in range(2*i + 1):
    print('*', end='')
  print()

for l in range(N-2, -1, -1):
  for m in range(N-l-1):
    print(' ', end='')
  for n in range(2*l + 1):
    print('*', end='')
  print()
```

```python

"""
Pyramid Pattern:

2. Modify the stair pattern to create a centered pyramid.


    *
   ***
  *****
 *******
*********

"""


N = 5
for i in range(N):
  for j in range(N-i-1):
    print(' ', end='')
  for k in range(2*i + 1):
    print('*', end='')
  print()
```

```python
"""
Hollow Square Pattern:

3. Create a hollow square where only the borders are *.

*****
*   *
*   *
*   *
*****

"""

N = 5
for i in range(N):
  print('*', end='')
print()
for j in range(N//2):
  print('*' + ' ' * (N-2) + '*')
for k in range(N):
  print('*', end='')
```

```python
# Program to check palindrome

word_1  = input().strip().lower()

check = word_1[::-1]
if check == word_1:
  print("Palindrome")
else:
  print("Not a Palindrome")
```

```python
# Program to count the number of vowels and print all vowels

vowels = ['A', 'E', 'I', 'O', 'U']
s = input().upper()
list_of_vowels = []
count = 0
for letter in s:
  if letter in vowels:
    list_of_vowels.append(letter)
    count += 1

print('The total number of vowels are:', count)
print('The list of vowels in the string are:', list_of_vowels)
```
